**4. Tuples in Python**



Tuples are immutable lists and cannot be changed in any way once it is created.

* Tuples are defined in the same way as lists.
* They are enclosed within parenthesis and not within square braces.
* Tuples are ordered, indexed collections of data.
* Similar to string indices, the first value in the tuple will have the index [0], the second value [1]
* Negative indices are counted from the end of the tuple, just like lists.
* Tuple also has the same structure where commas separate the values.
* Tuples can store duplicate values.
* Tuples allow you to store several data items including string, integer, float in one variable.

**Tuple Creation:**

````python
my_tuple = (1, 2, 3)
another_tuple = 4, 5, 6
````

**Accessing Tuple Elements:**

````python
my_tuple = (1, 2, 3)
print(my_tuple[0])   # Output: 1
print(my_tuple[2])   # Output: 3
````

**Tuple Unpacking:**

````python
my_tuple = (1, 2, 3)
a, b, c = my_tuple
print(a)   # Output: 1
print(b)   # Output: 2
print(c)   # Output: 3
````

**Immutable Nature:**

````python
my_tuple = (1, 2, [3, 4])
my_tuple[2][0] = 5   # Modifying a mutable object within a tuple
print(my_tuple)      # Output: (1, 2, [5, 4])
````

**Tuple Methods:**

````python
my_tuple = (1, 2, 2, 3, 3, 3)
print(my_tuple.count(2))     # Output: 2
print(my_tuple.index(3))     # Output: 3
````

**Tuple Concatenation and Repetition:**

````python
tuple1 = (1, 2)
tuple2 = (3, 4)
concatenated_tuple = tuple1 + tuple2
repeated_tuple = tuple1 * 3
````

**Tuple Packing and Unpacking:**

````python
# Tuple Packing
my_tuple = 1, 2, 3
print(my_tuple)    # Output: (1, 2, 3)

# Tuple Unpacking
a, b, c = my_tuple
print(a)   # Output: 1
print(b)   # Output: 2
print(c)   # Output: 3
````

**Immutable Keys in Dictionaries:**

````python
my_dict = {('John', 25): 'Engineer', ('Alice', 30): 'Manager'}
print(my_dict[('John', 25)])   # Output: 'Engineer'
````

**Returning Multiple Values from Functions:**

````python
def calculate_stats(numbers):
    minimum = min(numbers)
    maximum = max(numbers)
    mean = sum(numbers) / len(numbers)
    return minimum, maximum, mean

result = calculate_stats([1, 2, 3, 4, 5])
min_value, max_value, avg_value = result
print(min_value)   # Output: 1
print(max_value)   # Output: 5
print(avg_value)   # Output: 3.0
````

**5. Sets in Python**



* Set is one of 4 built-in data types in Python used to store collections of data including List, Tuple, and
Dictionary
* Sets are unordered, but you can remove items and add new items.
* Set elements are unique. Duplicate elements are not allowed.
* A set itself may be modified, but the elements contained in the set must be of an immutable type.
* Sets are used to store multiple items in a single variable.
* You can denote a set with a pair of curly brackets {}.


**Set Creation:**

````python
my_set = {1, 2, 3}
another_set = set([4, 5, 6])
````

**Set Elements:**

````python
my_set = {1, 2, 2, 3, 3, 3}
print(my_set)   # Output: {1, 2, 3}
````

**Modifying a Set:**

````python
my_set = {1, 2, 3}
my_set.add(4)        # Add a single element
my_set.update([5, 6])   # Add multiple elements
my_set.remove(3)     # Remove a specific element
my_set.discard(10)   # Remove an element (if present)
````

**Set Operations:**

````python
set1 = {1, 2, 3}
set2 = {3, 4, 5}
union = set1 | set2               # Union
intersection = set1 & set2        # Intersection
difference = set1 - set2          # Difference
symmetric_difference = set1 ^ set2   # Symmetric Difference
````

**Set Membership Testing:**

````python
my_set = {1, 2, 3, 4, 5}
print(3 in my_set)   # Output: True
print(6 in my_set)   # Output: False
````

**Set Size:**

````python
my_set = {1, 2, 3, 4, 5}
size = len(my_set)
print(size)   # Output: 5
````

**Set Methods:**

````python
set1 = {1, 2, 3}
set2 = {3, 4, 5}

set1.add(4)
print(set1)    # Output: {1, 2, 3, 4}

set1.remove(2)
print(set1)    # Output: {1, 3, 4}

union_set = set1.union(set2)
print(union_set)    # Output: {1, 3, 4, 5}

intersection_set = set1.intersection(set2)
print(intersection_set)    # Output: {3}

difference_set = set1.difference(set2)
print(difference_set)    # Output: {1, 4}

symmetric_difference_set = set1.symmetric_difference(set2)
print(symmetric_difference_set)    # Output: {1, 4, 5}
````

**6. Dictionaries in Python**



In Python, dictionaries are an unordered collection of key-value pairs, where each key is unique. Dictionaries are also known as associative arrays or hash maps.

* Dictionaries are used to store data values in key:value pairs.
* A dictionary is a collection which is ordered, changeable or mutable and do not allow duplicates.
* Dictionary items are ordered, changeable, and does not allow duplicates.
* Dictionary items are presented in key:value pairs, and can be referred to by using the key name.
* Dictionaries are changeable, meaning that we can change, add or remove items after the dictionary has been created.
* Dictionaries cannot have two items with the same key.
* A dictionary can nested and can contain another dictionary.

**Dictionary Creation:**

````python
my_dict = {'key1': 'value1', 'key2': 'value2', 'key3': 'value3'}
another_dict = dict(key1='value1', key2='value2', key3='value3')
````

**Accessing Values:**

````python
my_dict = {'name': 'John', 'age': 25, 'city': 'New York'}
print(my_dict['name'])    # Output: 'John'
print(my_dict['age'])     # Output: 25
````

**Modifying a Dictionary:**

````python
my_dict = {'name': 'John', 'age': 25, 'city': 'New York'}
my_dict['gender'] = 'Male'     # Add a new key-value pair
my_dict['age'] = 26            # Update the value of an existing key
del my_dict['city']            # Remove a key-value pair
````

**Dictionary Iteration:**

````python
my_dict = {'name': 'John', 'age': 25, 'city': 'New York'}

# Iterating over keys
for key in my_dict:
    print(key)   # Output: 'name', 'age', 'city'

# Iterating over values
for value in my_dict.values():
    print(value)   # Output: 'John', 25, 'New York'

# Iterating over items (key-value pairs)
for key, value in my_dict.items():
    print(key, value)   # Output: 'name' 'John', 'age' 25, 'city' 'New York'
````

**Dictionary Size:**

````python
my_dict = {'name': 'John', 'age': 25, 'city': 'New York'}
size = len(my_dict)
print(size)   # Output: 3
````

Comparison operations compare some value or operand and based on a condition, produce a Boolean. Python

Functions in Python are blocks of reusable code that perform specific tasks. They help in organizing code into logical chunks, making it easier to read, understand, and maintain. Functions also promote code reusability and modularity.

Functions can accept parameters, which are placeholders for values that are passed to the function during the function call. Parameters are specified within the function definition and can be used as variables within the function block.

**Variable Arguments:**

Python allows you to define functions that can accept a variable number of arguments. This is useful when you're unsure about the number of arguments that will be passed to the function. There are two types of variable arguments: positional arguments and keyword arguments.

````python
def calculate_sum(*args):
    total = 0
    for num in args:
        total += num
    return total

def print_info(**kwargs):
    for key, value in kwargs.items():
        print(key + ": " + value)
````

Variables defined inside a function have local scope and are accessible only within that function. Variables defined outside any function have global scope and can be accessed from any part of the program.

**Recursion:**

Recursion is a programming technique where a function calls itself to solve a problem. It can be useful for solving problems that can be broken down into smaller, similar sub-problems.

````python
def factorial(n):
    if n == 0:
        return 1
    else:
        return n * factorial(n - 1)
````

````python
factorial(5)
````

**Lambda Functions:**

Lambda functions, also known as anonymous functions, are small, one-line functions that don't require a def statement. They are commonly used with built-in functions like map(), filter(), and reduce().

````python
# Example 1: Doubling numbers using a lambda function
numbers = [1, 2, 3, 4, 5]
doubled_numbers = list(map(lambda x: x * 2, numbers))
print(doubled_numbers)  # Output: [2, 4, 6, 8, 10]
````

````python
# Example 2: Filtering even numbers using a lambda function
numbers = [1, 2, 3, 4, 5]
even_numbers = list(filter(lambda x: x % 2 == 0, numbers))
print(even_numbers)  # Output: [2, 4]
````

````python
# Example 3: Sorting a list of strings by their lengths using a lambda function
fruits = ["apple", "banana", "cherry", "date", "elderberry"]
sorted_fruits = sorted(fruits, key=lambda x: len(x))
print(sorted_fruits)  # Output: ['date', 'apple', 'banana', 'cherry', 'elderberry']
````

````python
# Example 4: Filtering numbers greater than a given threshold

numbers = [10, 25, 15, 30, 8, 12]
threshold = 20
filtered_numbers = list(filter(lambda x: x > threshold, numbers))
print(filtered_numbers)  # Output: [25, 30]
````

````python
# Example 5: Converting temperatures from Celsius to Fahrenheit

celsius_temperatures = [0, 10, 25, 32, 100]
fahrenheit_temperatures = list(map(lambda x: (9/5) * x + 32, celsius_temperatures))
print(fahrenheit_temperatures)  # Output: [32.0, 50.0, 77.0, 89.6, 212.0]
````

**Example of some Functions:**

````python
# Example 1: Function to calculate the factorial of a number
def factorial(n):
    if n == 0 or n == 1:
        return 1
    else:
        return n * factorial(n - 1)
````

````python
factorial(5)
````

````python
# Example 2: Function to check if a number is prime

def is_prime(n):
    if n < 2:
        return False
    for i in range(2, int(n ** 0.5) + 1):
        if n % i == 0:
            return False
    return True
````

````python
is_prime(11)
````

````python
# Example 3: Function to print the first n prime numbers
def print_n_primes(n):
    count = 0
    num = 2
    while count < n:
        if is_prime(num):
            print(num)
            count += 1
        num += 1
````

````python
print_n_primes(5)
````

In Python, you can read files using the built-in functions and methods provided by the language.

**CSV:**

```
import csv

file_path = 'path/to/file.csv'

# Method 1: Reading the entire CSV file
with open(file_path, 'r') as file:
    csv_data = file.read()
    print(csv_data)

# Method 2: Reading the CSV file line by line
with open(file_path, 'r') as file:
    csv_reader = csv.reader(file)
    for row in csv_reader:
        print(row)

# Method 3: Reading all rows into a list
with open(file_path, 'r') as file:
    csv_reader = csv.reader(file)
    rows = list(csv_reader)
    for row in rows:
        print(row)

```

for row in sheet.iter_rows(min_row=2, max_row=5, min_col=1, max_col=3, values_only=True):

````python
# Extracting Digits or Letters from a String:

text = "ABC123"
digits = ''.join([c for c in text if c.isdigit()])
letters = ''.join([c for c in text if c.isalpha()])
print(digits)   # Output: "123"
print(letters)  # Output: "ABC"
````

**18. Python Library**



In Python, a library (also referred to as a module) is a collection of pre-written code that provides specific functionality and can be used by other programs. A library typically consists of reusable functions, classes, and variables that can be imported into your code to perform specific tasks or extend the capabilities of the Python language.

Libraries serve the purpose of code reusability, allowing developers to leverage existing code instead of writing everything from scratch. They encapsulate complex operations or algorithms into pre-built functions, making it easier to use and maintain.

Python has a vast ecosystem of libraries and modules that extend its capabilities for various purposes. Here are some commonly used libraries in Python:

**Data Science:**

* NumPy: A fundamental library for numerical computing in Python. It provides support for large, multi-dimensional arrays and matrices, along with a collection of mathematical functions to operate on these arrays efficiently.

* Pandas: A versatile library for data manipulation and analysis. It offers data structures like DataFrames, which allow for easy handling and manipulation of structured data. Pandas provides functionalities for data cleaning, filtering, merging, reshaping, and more.

* Matplotlib: A plotting library that allows you to create static, animated, and interactive visualizations in Python. Matplotlib provides a wide range of plot types and customization options to visualize data effectively.

* Seaborn: Built on top of Matplotlib, Seaborn provides a higher-level interface for creating informative and visually appealing statistical graphics. It simplifies the process of creating complex statistical visualizations and offers additional statistical plotting functions.

* SciPy: A library for scientific and technical computing. It includes modules for optimization, integration, interpolation, signal and image processing, linear algebra, statistics, and more. SciPy complements NumPy and provides additional functionality for scientific computing tasks.

* Scikit-learn: A machine learning library that provides tools for data mining, data preprocessing, classification, regression, clustering, model selection, and evaluation. Scikit-learn offers a wide range of algorithms and utilities for machine learning tasks.

* TensorFlow: An open-source library for deep learning and neural networks. TensorFlow provides a flexible and comprehensive ecosystem for building and deploying machine learning models. It offers high-level APIs like Keras for building neural networks efficiently.

* PyTorch: Another popular deep learning library that emphasizes flexibility and dynamic computation graphs. PyTorch provides tools for building and training neural networks, and it has gained popularity for its ease of use and extensive community support.

* Statsmodels: A library for statistical modeling and estimation. Statsmodels provides a wide range of statistical models, regression analysis, time series analysis, hypothesis testing, and more. It integrates well with NumPy and Pandas.

* Jupyter Notebook: Although not a library, Jupyter Notebook is an interactive computing environment that allows you to create and share documents that contain code, visualizations, and narrative text. It is widely used in data science for exploratory data analysis, prototyping, and creating data-driven reports.

* Theano: Theano is a library that allows for efficient mathematical operations and expression optimization, especially for deep learning. While its development has slowed down, it is still used in some legacy projects.

* XGBoost: XGBoost is an optimized gradient boosting library that provides high-performance implementations of gradient boosting algorithms. It is known for its efficiency and effectiveness in handling structured data for both classification and regression tasks.

* NLTK: The Natural Language Toolkit (NLTK) is a library for working with human language data. It provides tools and resources for tasks like tokenization, stemming, part-of-speech tagging, named entity recognition, and more. NLTK is widely used in natural language processing (NLP) applications.

**Web Development:**

* Django: A full-featured web framework that follows the model-view-controller (MVC) architectural pattern. Django provides a robust set of tools for building web applications, including an ORM (Object-Relational Mapping) for database management, a templating engine, and built-in authentication and admin interfaces.

* Flask: A lightweight and flexible microframework for web development. Flask allows you to build web applications quickly and with minimal boilerplate code. It provides URL routing, template support, and extension libraries for database integration, form handling, and more.

* FastAPI: A modern, high-performance web framework for building APIs with Python 3.7+ type hints. FastAPI is known for its speed and ease of use. It leverages Python's async and await keywords for efficient and scalable code execution.

* Bottle: A simple and lightweight web framework that is ideal for small projects and APIs. Bottle has minimal dependencies and a small codebase, making it easy to learn and use. It includes built-in support for templates, routing, and request handling.

* Pyramid: A flexible and scalable web framework that can be used for projects of all sizes. Pyramid follows a minimalist approach, providing only the essentials and allowing developers to choose and integrate additional components as needed.

* CherryPy: A minimalist web framework that focuses on simplicity and ease of use. CherryPy allows you to build web applications and services with minimal effort. It includes its own HTTP server and supports a variety of deployment options.

* Tornado: An asynchronous web framework that excels in handling high-traffic and real-time applications. Tornado is known for its non-blocking and scalable nature. It includes a built-in web server and supports WebSockets and other asynchronous features.

* SQLAlchemy: A powerful SQL toolkit and Object-Relational Mapping (ORM) library. SQLAlchemy simplifies database access and manipulation in web applications by providing a Pythonic interface to interact with databases. It supports multiple database backends and provides high-level ORM features.

* Gunicorn: A lightweight and highly-scalable HTTP server that is commonly used for deploying Python web applications. Gunicorn can serve web applications built with various frameworks, including Django, Flask, and Pyramid.

* Celery: A distributed task queue library that allows you to run asynchronous and scheduled tasks in your web applications. Celery is often used for background processing, handling long-running tasks, and offloading computationally intensive operations.

* Pygame: Pygame is a widely-used library for creating 2D games in Python. It provides functionality for handling graphics, sound, input, and more. Pygame is relatively easy to learn and is suitable for beginners in game development.

* Godot: While not a Python library, Godot is an open-source game engine that has its own scripting language similar to Python. It provides a visual editor and supports both 2D and 3D game development. Godot has gained popularity for its user-friendly interface and extensive feature set.
